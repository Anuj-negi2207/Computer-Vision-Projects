```

import cv2 
import numpy as np 
   
      
# Read the image 
img = [4, 2, 3, 6, 2, 3, 4, 5, 2, 3, 4, 5, 5, 1, 5]
  
# Obtain number of rows and columns  
# of the image 
m = 1 
n = 15 
   
# Develop Averaging filter(3, 3) mask 
mask = [1, 0, 1]
   
# Convolve the 3X3 mask over the image  
img_new = [0 for i in range(n)] 
  
 
for i in range(1, n-1): 
  temp = img[i-1]*mask[0] + img[i]*mask[1] + img[i+1]*mask[2] 
  img_new[i]= temp 
          
print(img_new)
```
