```
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```
def Process(img, save):
  source = cv2.imread(img, cv2.IMREAD_COLOR)
  #Remove noise by applying Gaussian Blur:
  source1 = cv2.GaussianBlur(source, (3,3), 0)

  #Here, first argument is the image. Second defines the kernel size. 
  #Convert to grayscale:
  source_gray = cv2.cvtColor(source1, cv2.COLOR_BGR2GRAY)
  #Laplacian filter
  dest = cv2.Laplacian(source_gray, cv2.CV_16S, ksize = 3)
  abs_dest = cv2.convertScaleAbs(dest)
  
  #OUTPUT
  print("Original Image: \n")
  cv2_imshow(source)
  print("\n\nAfter applying Laplacian Filter: \n")
  cv2_imshow(abs_dest)
  print("\n\nUSING cmap = gray:\n")
  plt.imshow(abs_dest, cmap = "gray")

  #saving the file
  cv2.imwrite(save, abs_dest)
```

# **MAIN**

```
img = "1.jpg"
save = "Saved1.jpg"
Process(img, save)
```

```
img = "2.jpg"
save = "Saved2.jpg"
Process(img, save)
```

```
img = "3.jpg"
save = "Saved3.jpg"
Process(img, save)
```
