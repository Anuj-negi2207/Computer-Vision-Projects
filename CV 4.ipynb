```
import cv2
from google.colab.patches import cv2_imshow
import numpy as np 
import matplotlib.pyplot as plt
```

```
def Process(img, save):
  source = cv2.imread(img)
  gray = cv2.cvtColor(source, cv2.COLOR_BGR2GRAY)

  img_gaussian = cv2.GaussianBlur(gray, (3,3), 0)

  #CANNY
  img_canny = cv2.Canny(source, 100, 200)

  #SOBEL
  img_sobelx = cv2.Sobel(img_gaussian, cv2.CV_8U, 1, 0, ksize = 5)
  img_sobely = cv2.Sobel(img_gaussian, cv2.CV_8U, 0, 1, ksize = 5)

  img_sobel = img_sobelx + img_sobely


  #PREWITT
  kernelX = np.array([[1,1,1], [0,0,0], [-1,-1,-1]])
  kernelY = np.transpose(kernelX*(-1))

  img_prewittX = cv2.filter2D(img_gaussian, -1, kernelX)
  img_prewittY = cv2.filter2D(img_gaussian, -1, kernelY)
  img_prewitt = img_prewittX + img_prewittY

  print("Original Image:\n")
  cv2_imshow(source)

  print("\n\nCanny:\n")
  cv2_imshow(img_canny)
  cv2.imwrite("Canny "+ save, img_canny)

  print("\n\nSobel X:\n")
  cv2_imshow(img_sobelx)
  cv2.imwrite("Sobel X " + save, img_sobelx)

  print("\n\nSobel Y:\n")
  cv2_imshow(img_sobely)
  cv2.imwrite("Sobel Y " + save, img_sobely)
  
  print("\n\nSobel:\n")
  cv2_imshow(img_sobel)
  cv2.imwrite("Sobel " + save, img_sobel)
  
  print("\n\nPrewitt X:\n")
  cv2_imshow(img_prewittX)
  cv2.imwrite("Prewitt X " + save, img_prewittX)
  
  print("\n\nPrewitt Y:\n")
  cv2_imshow(img_prewittY)
  cv2.imwrite("Prewitt Y " + save, img_prewittY)
  
  print("\n\nPrewitt:\n")
  cv2_imshow(img_prewitt)
  cv2.imwrite("Prewitt " + save, img_prewitt)
```

# **MAIN**

```
img = "1.jpg"
save = "Saved1.jpg"
Process(img, save)
```

```
img = "2.jpg"
save = "Saved2.jpg"
Process(img, save)
```
